```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pn = pd.read_csv("200MHzPhaseNoise.dat", delimiter=';', skiprows=78, header=None, 
                 usecols=[0,1], names=['freq', 'dBc/Hz'])
pnref = pd.read_csv("200MHzPhaseNoise_ref.dat", delimiter=';', skiprows=78, header=None, 
                 usecols=[0,1], names=['freq', 'dBc/Hz'])

plt.semilogx(pn['freq'], pn['dBc/Hz'], pnref['freq'], pnref['dBc/Hz'])
plt.title('Phase Noise @ 200MHz (40MHz ref. oscillator)')
plt.ylabel('dBc/Hz')
plt.xlabel('Freq. offset (Hz)')
plt.savefig('phase_noise.pdf')

print("Phase noise @ 100 Hz: %f dBc/Hz" % pn.loc[100]['dBc/Hz'])
print("Phase noise @ 1 kHz: %f dBc/Hz" % pn.loc[1e3]['dBc/Hz'])
print("Phase noise @ 10 kHz: %f dBc/Hz" %  pn.loc[pn.freq >= 1e4].iloc[0]['dBc/Hz'])
print("Phase noise @ 100 kHz: %f dBc/Hz" %  pn.loc[pn.freq >= 1e5].iloc[0]['dBc/Hz'])
print("Phase noise @ 1 MHz: %f dBc/Hz" %  pn.loc[pn.freq >= 1e6].iloc[0]['dBc/Hz'])
```

```python
spectrum = pd.read_csv("200MHzSpectrum.dat", delimiter=';', skiprows=29, header=None, 
                 usecols=[0,1], names=['freq', 'dBm'])

plt.plot(spectrum.freq, spectrum.dBm)
plt.ylabel('Power (dBm)')
plt.xlabel('Frequency (Hz)')
plt.title('200MHz output, 40MHz ref. oscillator')
plt.savefig('spectrum.pdf')
```
